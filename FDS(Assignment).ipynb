```python
import pandas as pd
```

```python
#Q1 Create a DataFrame named financial_data that contains quarterly financial metrics for the years 2021 and 2022. The metrics include "Sales" and"Expenses," which are further divided into "Online" and "In-Store"channels.
financial_data = pd.DataFrame({
    'Year': [2021, 2021, 2021, 2021, 2022, 2022, 2022, 2022],
    'Quarter': ['Q1', 'Q2', 'Q3', 'Q4', 'Q1', 'Q2', 'Q3', 'Q4'],
    'Sales_Online': [15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000],
    'Sales_InStore': [30000, 32000, 33000, 34000, 35000, 36000, 37000, 38000],
    'Expenses_Online': [5000, 5500, 6000, 6500, 7000, 7500, 8000, 8500],
    'Expenses_InStore': [10000, 10500, 11000, 11500, 12000, 12500, 13000, 13500]
})
financial_data
```

```python
#(a)What are the online sales for Q1 2022?
online_sales_q1_2022 = financial_data[(financial_data['Year'] == 2022) & 
                                      (financial_data['Quarter'] == 'Q1')]['Sales_Online'].values[0]
online_sales_q1_2022
```

```python
#(b)How much were the in-store expenses in Q3 2021?
in_store_expenses_q3_2021 = financial_data[(financial_data['Year'] == 2021) & 
                                           (financial_data['Quarter'] == 'Q3')]['Expenses_InStore'].values[0]
in_store_expenses_q3_2021
```

```python
#(c)What are the total sales for Q4 2021?
total_sales_q4_2021 = financial_data[(financial_data['Year'] == 2021) & 
                                     (financial_data['Quarter'] == 'Q4')][['Sales_Online', 'Sales_InStore']].sum(axis=1).values[0]
total_sales_q4_2021
```

```python
#(d)What were the expenses for the online channel in 2022?
online_expenses_2022 = financial_data[financial_data['Year'] == 2022]['Expenses_Online'].sum()
online_expenses_2022
```

```python
#(e) Retrieve sales data for Q2 2021.
sales_q2_2021 = financial_data[(financial_data['Year'] == 2021) & 
                               (financial_data['Quarter'] == 'Q2')][['Sales_Online', 'Sales_InStore']].values[0]
sales_q2_2021
```

```python
#(f)What is the maximum expense recorded in 2022?
max_expense_2022 = financial_data[financial_data['Year'] == 2022][['Expenses_Online', 'Expenses_InStore']].max()
max_expense_2022
```

```python
#Q2 Stack a DataFrame of Store, Product, and Sales.
DataFrame= pd.DataFrame( {
    'Store': ['Store1', 'Store2', 'Store3'],
    'Product': ['ProductA', 'ProductB', 'ProductC'],
    'Sales': [100, 200, 150]
})
DataFrame
```

```python
stacked_df = DataFrame.set_index(['Store', 'Product']).stack().reset_index()
stacked_df.columns = ['Store', 'Product', 'Variable', 'Value']
stacked_df
```

```python
#Q3 Unstack it to show sales by store.
unstacked_df = DataFrame.pivot(index='Product', columns='Store', values='Sales')
unstacked_df
```

```python
#Q10 Create a multi-index using Category and Product in a product DataFrame.Then, average the levels in the column.
Product =pd.DataFrame ({
    'Category': ['Electronics', 'Electronics', 'Furniture', 'Furniture', 'Electronics', 'Furniture'],
    'Product': ['Laptop', 'Smartphone', 'Table', 'Chair', 'Tablet', 'Sofa'],
    'Price': [1000, 600, 150, 200, 300, 400],
    'Quantity': [5, 10, 3, 4, 7, 2]
})
Product
```

```python
Product.reset_index()
```

```python
avg_category = Product.groupby(level='Category').mean()
avg_category
```
